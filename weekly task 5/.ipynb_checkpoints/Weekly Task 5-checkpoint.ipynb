# Question 1

Using the dataset "us-cities-top-1k-multi-year.csv":
1. Create a line graph of total population over time for the states: NY, PA, CT, NJ (10 pts)
2. Be sure to include a title of the graph, labels on your axis, and a legend. (5 pts)
3. Specify colors to be used in your code. (5 pts)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid") ##set the style to a white grid

df = pd.read_csv('us-cities-top-1k-multi-year.csv') ##load the cities dataset

plt.figure(figsize=(8, 6))##create a single graph

##filtering using or method
q1_data = df[(df['State'] == 'New York') | (df['State'] == 'Pennsylvania') | 
             (df['State'] == 'Connecticut') | (df['State'] == 'New Jersey')]
state_pop_over_time = q1_data.groupby(['year', 'State'])['Population'].sum().reset_index() ##groups the data by year and state, and adds up the total population for each

custom_colors = {'New York': 'blue', 'Pennsylvania': 'green', 'Connecticut': 'red', 'New Jersey': 'purple'} ##custom colors


sns.lineplot(data=state_pop_over_time, x='year', y='Population', hue='State', palette=custom_colors) ##draws the line graph using our grouped data

ax1.set_title('Total Population Over Time by State')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Population')
ax1.ticklabel_format(axis='y', style='plain')

plt.show()
```

# Question 2

Using the dataset "us-cities-top-1k-multi-year.csv":
1. Create a Bar Graph of the total population for 5 cities in the state of New York comparing 2014 years and 2018 years (15 pts)
2. Be sure to include a title of the graph, labels on your axis, and a legend. (5 pts)
3. Combine the graph in question 1 and question 2 in one figure.

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(15, 6)) ##creating the space to combine the graphs

##GRAPH FROM QUESTION 1
q1_data = df[(df['State'] == 'New York') | (df['State'] == 'Pennsylvania') | 
             (df['State'] == 'Connecticut') | (df['State'] == 'New Jersey')]

state_pop_over_time = q1_data.groupby(['year', 'State'])['Population'].sum().reset_index()

custom_colors = {'New York': 'blue', 'Pennsylvania': 'green', 'Connecticut': 'red', 'New Jersey': 'purple'} 

ax1 = plt.subplot(1, 2, 1) ##assigns the left box to the variable ax1
sns.lineplot(data=state_pop_over_time, x='year', y='Population', hue='State', palette=custom_colors, ax=ax1)

ax1.set_title('Total Population Over Time by State')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Population')
ax1.ticklabel_format(axis='y', style='plain')


##filtering w/ the | operator again
q2_data = df[(df['State'] == 'New York') & 
             ((df['City'] == 'New York') | (df['City'] == 'Buffalo') | (df['City'] == 'Rochester') | 
              (df['City'] == 'Yonkers') | (df['City'] == 'Syracuse')) & 
             ((df['year'] == 2014) | (df['year'] == 2018))]


##sassigning to ax2 
ax2 = plt.subplot(1, 2, 2)
sns.barplot(data=q2_data, x='City', y='Population', hue='year', ax=ax2, palette='Set2')
ax2.set_title('Population Comparison: 2014 vs 2018')
ax2.set_xlabel('City Name')
ax2.set_ylabel('Population')
ax2.ticklabel_format(axis='y', style='plain') 

fig.tight_layout() ##adjusts the spacing between the two graphs
plt.show()
```

## Question 3

import the data "job-automation-probability.csv"
1. A Histogram of the variable 'probability'
2. A scatter plot of 'Average annual wage' and 'numbEmployed'
3. Put both of these graphs in one figure
4. Using item 3, change the ticks of Number Employed axes so that it is in logarithmic axis (hint: it's an option) 
5. Adding onto item 4, add a title and axis labels
6. Save your figure as a png

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid") ##sets the style of the graph to have a white background with gridlines

jobs_df = pd.read_csv('job-automation-probability.csv') ##loads the job automation dataset

fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(14, 5)) ##creates a container with 1 row and 2 columns so we can put both graphs next to each other

ax1 = plt.subplot(1, 2, 1) ##assigns the first box to the variable ax1

sns.histplot(data=jobs_df, x='probability', bins=20, color='teal', ax=ax1) ##draws a histogram using the probability column

ax1.set_title('Distribution of Job Automation Probability')
ax1.set_xlabel('Probability of Automation')
ax1.set_ylabel('Frequency (Number of Jobs)')

ax2 = plt.subplot(1, 2, 2) ##assigns the second space to the variable ax2

sns.scatterplot(data=jobs_df, x='numbEmployed', y='Average annual wage', color='darkorange', ax=ax2)

ax2.set_xscale('log') ##changes the x axis to a log scale

ax2.set_title('Average Annual Wage vs. Number Employed')
ax2.set_xlabel('Number Employed (Logarithmic Scale)')
ax2.set_ylabel('Average Annual Wage ($)')

##so the words and labels don't overlap
fig.tight_layout() 

##saves figure as a png file
plt.savefig('automation_graphs.png', dpi=300, bbox_inches='tight')

plt.show()
```
